```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('../depth5.info', 'rb') as f:
    arr = np.frombuffer(f.read(), dtype=np.uint8).reshape((-1, 2))
    df = pd.DataFrame(arr, columns=['depth', 'parity'])
```

```python
import seaborn as sns
```

```python
plt.hist(df['parity'], density=True)
plt.show()
```

```python
def hist_group_by(x, group_by, df):
    for (group, subset) in df.groupby(group_by):
        plt.title(f'{group_by} = {group}')
        plt.hist(subset[x], bins=np.arange(0, 41, 1))
        plt.xlim(0, 40)
        plt.show()
```

```python
hist_group_by('parity', ['depth'], df)
```
